```python
def year_month(function):
    def wrapper():
        #print(function())
        df=function()
        #print(type(df))
        df.loc[:,'Year']=df.Date.dt.year
        df.loc[:,'month']=df.Date.dt.month
        return df
    return wrapper


```

```python
@year_month
def get_file():
    df=pd.read_csv(r'C:\Users\dlkdxp8\Downloads\AirQualityUCI\AirQualityUCI.csv',delimiter=';')
    df_skipna=df[~df['Date'].isna()]
    df_skipna.loc[:,'Date']=pd.to_datetime(df_skipna['Date'])
    return df_skipna
    
    
```
